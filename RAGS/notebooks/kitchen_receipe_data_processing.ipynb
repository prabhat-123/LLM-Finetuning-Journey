```python
import os
import json
import zipfile
import warnings
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

## Loading the csv dataset in a pandas dataframe

```python
path_to_dataset = os.path.join(os.path.dirname(os.getcwd()), 'receipe_data', 'IndianFoodDatasetCSV.csv')
```

```python
df = pd.read_csv(path_to_dataset)
print(df.shape)
df.head()
```

```python
filter_df = df[['TranslatedRecipeName', 'TranslatedIngredients',
                'PrepTimeInMins', 'CookTimeInMins',
                'TotalTimeInMins', 'Servings', 'Cuisine',
                'Course', 'Diet', 'TranslatedInstructions']]
print(filter_df.shape)
filter_df.head()
```

## Dropping Duplicate Recipe Names

```python
filter_df = filter_df[~filter_df['TranslatedRecipeName'].duplicated()].reset_index(drop = True)
print(filter_df.shape)
```

```python
filter_df.head()
```

```python
filter_df['len_receipe'] = filter_df['TranslatedInstructions'].str.split(' ').str.len()
```

```python
filter_df['len_receipe'].describe()
```

```python
filter_df['len_receipe'].quantile(0.8)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def create_receipe_instructions(row):
  receipe_question = "How can we make " + row['TranslatedRecipeName'] + "? "
  receipe_instructions = receipe_question + row['TranslatedInstructions']
  return receipe_instructions
```

```python
filter_df['receipe_instructions'] = filter_df.apply(create_receipe_instructions,
                                                    axis=1)
```

```python
import pycld2 as cld2
```

```python
def detect_language(row):
  _, _, _, detected_language = cld2.detect(row['receipe_instructions'],
                                         returnVectors=True)
  return detected_language
```

```python
filter_df['detected_language'] = filter_df.apply(detect_language,
                                                 axis=1)
```

```python
def count_unique_language(row):
  return len(row)
```

```python
filter_df['no_of_language'] = filter_df['detected_language'].apply(count_unique_language)
```

```python
filter_df = filter_df.loc[filter_df['no_of_language']==1].reset_index(drop=True)
print(filter_df.shape)
```

```python
filter_df.drop(['len_receipe', 'detected_language', 'no_of_language'],
               axis=1, inplace = True)
```

```python
filter_df.shape
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model =  SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def create_embedding(row, model):
  embeddings = model.encode(row)
  return embeddings
```

```python
filter_df['receipe_instructions_embeddings'] = filter_df['receipe_instructions'].apply(lambda x: create_embedding(x, model))
```

```python
output_data_path = os.path.join(os.path.dirname(os.getcwd()), 'receipe_data', 'indian_food_receipe.csv')
```

```python
filter_df.to_csv(output_data_path, index=False)
```
